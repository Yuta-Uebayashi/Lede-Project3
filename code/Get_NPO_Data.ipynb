## Scrape HTML from NPO Portal

```python
import os
import random
from playwright.async_api import async_playwright,expect
```

```python
#ディレクトリを作成
os.makedirs('/Users/yuta-uebayashi/Desktop/Lede課題一覧/Lede個人プロジェクト/project3/npo_data', exist_ok=True)
```

```python
async def open_browser(headless=False):
     playwright = await async_playwright().start()
     browser = await playwright.chromium.launch(headless=headless)
     context = await browser.new_context()
     page = await browser.new_page()
     return browser, page
```

```python
browser, page = await open_browser()
```

```python
url = 'https://www.npo-homepage.go.jp/npoportal/index'
```

```python
await page.goto(url)
```

```python
#「災害救援」で条件付き検索
xpath_search_option = '//a[text()="災害救援"]'
```

```python
filter = page.locator(xpath_search_option)
```

```python
await filter.click()
```

```python
#セーブ
source = await page.content()
with open('/Users/yuta-uebayashi/Desktop/Lede課題一覧/Lede個人プロジェクト/project3/npo_data/all_NPOs.html', 'w') as f:
    f.write(source)
```

```python
#「次ページ」を探してリピート
xpath_next = '//a[@rel="next"]'
```

```python
#「結果」を確認
xpath_result = '//p[@class="pagination-info"]'
```

```python
next_button = page.locator(xpath_next).nth(1)
```

```python
await expect(next_button).to_be_visible()
```

```python
await next_button.click()
```

```python
# 最初のページを保存
i = 1
source = await page.content()
file_path = f'/Users/yuta-uebayashi/Desktop/Lede課題一覧/Lede個人プロジェクト/project3/npo_data/page_{i}.html'
with open(file_path, 'w') as f:
    f.write(source)

# 次のページ以降を繰り返し取得
collect = True

while collect:
    try:
        await expect(page.locator(xpath_next).first).to_be_visible(timeout=1000)
        next_button = page.locator(xpath_next).first
        await next_button.click()

        await page.locator(xpath_result).is_visible()

        i += 1  # ← ページ番号を更新（1ページ目は保存済み）

        source = await page.content()
        file_path = f'/Users/yuta-uebayashi/Desktop/Lede課題一覧/Lede個人プロジェクト/project3/npo_data/page_{i}.html'
        with open(file_path, 'w') as f:
            f.write(source)

    except Exception as e:
        print(f"停止しました: {e}")
        collect = False
        break
```

```python
import pandas as pd
import glob

# ファイルのパスをまとめて取得
files = sorted(glob.glob('/Users/yuta-uebayashi/Desktop/Lede課題一覧/Lede個人プロジェクト/project3/npo_data/page*.html'))

# データを格納するリスト
all_tables = []

for fn in files:
    tables = pd.read_html(fn)  # 各HTML内のすべてのテーブルを抽出
    if tables:  # 少なくとも1つあれば
        df = tables[0]  # 通常は1ページに1つだけ
        all_tables.append(df)

# すべてのテーブルを1つのDataFrameに結合
full_df = pd.concat(all_tables, ignore_index=True)

# 確認
print(full_df.shape)
display(full_df.head())
```

```python
# CSVに保存
full_df.to_csv('/Users/yuta-uebayashi/Desktop/Lede課題一覧/Lede個人プロジェクト/project3/npo_data/all_NPOs_pandas.csv', index=False)
```
