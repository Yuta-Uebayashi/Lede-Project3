### This notebook analyzes the historical trends in the number of elementary schools and students in Japan, starting from 1948. I clean the data, convert it to numeric types, and visualize key trends.

```python
import pandas as pd
df = pd.read_csv("/Users/yuta-uebayashi/Desktop/Lede課題一覧/Lede個人プロジェクト/project3/school_data/Number of Elementary Schools and Elementary Students.csv")
df.head()
```

```python
# カンマを除去し、int型に変換
df["Number of Schools"] = df["Number of Schools"].str.replace(",", "").astype(int)
df["Elementary Students"] = df["Elementary Students"].str.replace(",", "").astype(int)

# 確認
df.dtypes
df.head()
```

```python
import matplotlib.pyplot as plt

# 図と軸の作成
fig, ax1 = plt.subplots(figsize=(12, 6))

# 左Y軸：折れ線グラフ（学校数）
ax1.plot(df["year"], df["Number of Schools"], color="tab:blue", label="Number of Schools")
ax1.set_xlabel("Year")
ax1.set_ylabel("Number of Schools", color="tab:blue")
ax1.tick_params(axis='y', labelcolor="tab:blue")

# 右Y軸：棒グラフ（児童数）
ax2 = ax1.twinx()
ax2.bar(df["year"], df["Elementary Students"], color="tab:green", alpha=0.4, label="Elementary Students")
ax2.set_ylabel("Elementary Students", color="tab:green")
ax2.tick_params(axis='y', labelcolor="tab:green")

# タイトル・グリッド
plt.title("Number of Elementary Schools (Line) and Students (Bar) in Japan (1948–)")
ax1.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 整えたデータをCSVに出力
df.to_csv("/Users/yuta-uebayashi/Desktop/cleaned_elementary_data.csv", index=False)
```
